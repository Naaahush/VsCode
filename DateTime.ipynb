```python
import datetime
```

```python
d = datetime.date(2023,2,17) #yyyy-mm-dd
print(d)

# zero padding is not allowed in datetime
```

```python
#get today's date
tday = datetime.date.today()

print(tday)
```

```python
# just want to grab year,month or day

tday = datetime.date.today()

print(tday.year)

tday = datetime.date.today()

print(tday.month)

tday = datetime.date.today()

print(tday.day)

tday = datetime.date.today()

print(tday.weekday())

print(tday.isoweekday())

# for Weekday : Monday is 0 and Sunday is 6

# for isoweekday : Monday is 1 and Sunday is 7
```

```python
#timedelta --> they are the difference between two dates and times
# they are preffered for date and time operations

# create time delta which will be one week away
t_Delta = datetime.timedelta(days=7)

#print what date will be from 7 days from now

print(tday + t_Delta)
```

```python
print(tday - t_Delta)
```

```python
##How many days are there till my birthday this year

bday=datetime.date(2023,9,4)

#timedelta
#time till birthday is my bday minus the current date
till_bday = bday - tday
print(till_bday)

print(till_bday.days)

print(till_bday.total_seconds())
```

```python
# datetime.time we work with hours,minutes,seconds,milliseconds
# this does not have timezone information
# we can do that by tzinfo
t = datetime.time(9,30,45,10000) #(hours,minutes,seconds,milliseconds)
print(t)

print(t.hour)
```

```python
#datetime.datetime will give us everything

dt = datetime.datetime(2023,2,17,6,15,30,10000)
print(dt)
print(dt.date())
print(dt.time())
print(dt.year)
```

```python
time_delta = datetime.timedelta(days=7)

print(dt+time_delta)
```

```python
today= datetime.datetime.now()
today
```

```python
#for timedelta you can also change your duration in hours
time_delta = datetime.timedelta(hours=16)

print(dt+time_delta)

# we are adding 16 hours to today's date
```

```python
# working with timezones
# we work with pytz library for timezones
# pytz bringsn the timezone database (also known as the Olson database) to python and its usage is recommended
# to install --> pip install pytz
```

```python
import pytz
```

```python
#creating timezone aware data

dt = datetime.datetime(2023,2,17,6,15,45,tzinfo=pytz.UTC)
print(dt)

# datetime has +00 , which is a UTC offset
```

```python
dt_now = datetime.datetime.now(tz=pytz.UTC)
print(dt_now)
# this is easier and preferred option
```
