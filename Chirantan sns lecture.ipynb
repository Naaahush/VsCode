```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/VsCode/mtcars.csv')
df.head()
```

```python
df.info()
```

```python
df.shape

#print(np.shape(df))
```

```python
#barplot
#plotting no of cylinders against number of mpg(mileage per gallon) 
# the line in the top middle of the bars are the errors of the perticular bar


result = sns.barplot(data=df, x='cyl', y='mpg')
#plt.show()
```

Observation: As the number of cylinders increases the mileage is decreasing.

Number of Cylinders is 'negatively correlated' to the Mileage per gallon.

```python
#This is the same graph with color changed.
result = sns.barplot(data=df, x='cyl', y='mpg',color='purple')
```

```python
#Plotting bargraph - am(Automatic/Manual) with mpg(Miles per gallon)

result = sns.barplot(data=df, x='am', y='mpg',palette='Set2')
```

## Obersvation:

The number of miles per gallon is higher in manual(1) than compared to automatic(0)

```python
#countplot 
#returns count for each cstegory


result = sns.countplot(data=df,x='cyl', palette='tab10')
#There is not y axis here, since it is just showing the count 
```

# Observation:-
Most of the cars present in the data has 8 cylinders

```python
result=sns.countplot(data=df, y='gear', palette='Set2')
plt.show()
```

# Observation:-
Most cars have 3 gears.

```python
#Group based countplot where we group the nnumber of cylinders with the number of gears present in the car

sns.countplot(data=df, x='gear',hue='cyl',palette='tab10')
```

# Observation:-
### Most of the cars with 3 gears have 8 cylinders.
### None of the cars having 4 gears have 8 cylinders.
### The number of cars having 5 gears and 8 cylinders is almost same as the number of cars having 5 gears and 4 cylinders.

```python
#distplot :- creates a distribution of continuos data: age,height,weight

sns.displot(df.mpg, bins=10,color='r') #mention the column name directly using df.mpg
```

# Observation:-
MPG has pretty much normal distribution

```python
#catplot
#the jitter parameter controls the magnitude of the jitter 
tips =sns.load_dataset('tips')

sns.catplot(tips,x='day',y='total_bill',jitter=False,hue='day')
```

```python
print(tips)
```

```python
sns.catplot(tips,x='smoker',y='tip',hue='sex',order=['No','Yes'])
```

```python
#distplot

sns.displot(data=tips,x='size',hue='smoker')
```

# Observation:-
Among the people who came in groups of 2, around 70% are smokers.

Among the group of 6, none of them smoke.

```python
sns.displot(data=tips,x='total_bill',col='time',hue='sex')
```

# observations:-

Highest tips are collected are during dinner.

As the total bill increases the tipping amount increases in the dinner time as compared to lunch time.

Male customers tipped higher in the dinner than the female customers. 

```python
#barplot 

sns.barplot(tips,x='total_bill',y='sex',palette='Set2',hue='smoker')
```

# observations:-

Male customers have a greater spend on average than female customers.

Total bill is higher from the smoker male customers than non smoker males.

Total bill amount is relatively unaffected in the both category of female customers.

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="smoker",
    kind="swarm", col="time", aspect=.7)
```

# observations:-

Most amount of total_bill is collected on thursday.

Saturday and Sunday dinners are the peak time for maximum billing.

Non smokers are on an avergae making a total bill around 10-30 dollars on weekends.
